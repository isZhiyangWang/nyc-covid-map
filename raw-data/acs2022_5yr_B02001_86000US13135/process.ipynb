```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_csv("acs2022_5yr_B02001_86000US13135.csv")
```

```python
df
```

```python
race_columns = ['B02001001', 'B02001002', 'B02001003', 'B02001004', 'B02001005', 
                'B02001006', 'B02001007', 'B02001008', 'B02001009', 'B02001010']
```

```python
json_data = []
for index, row in df.iterrows():
    entry = {
        "zipcode": row['name'],
        "race": [row[col] for col in race_columns]
    }
    json_data.append(entry)


json_output = json.dumps(json_data, indent=2)


print(json_output)
with open('output.json', 'w') as f:
    json.dump(json_data, f, indent=2)
```

```python

with open('race.json', 'r') as race_file:
    race_data = json.load(race_file)

with open('income.json', 'r') as income_file:
    income_data = json.load(income_file)


race_dict = {item['zipcode']: item for item in race_data}


merged_data = []
for income_entry in income_data:
    zipcode = income_entry['zipcode']
    if zipcode in race_dict:
        merged_entry = {**race_dict[zipcode], **income_entry}  # Merge dictionaries
        merged_data.append(merged_entry)


with open('merged_output.json', 'w') as merged_file:
    json.dump(merged_data, merged_file, indent=2)


```

```python
json_data_dict = {}
for index, row in df.iterrows():
    json_data_dict[row['name']] = [row[col] for col in race_columns]


json_output_dict = json.dumps(json_data_dict, indent=2)


print(json_output_dict)

with open('race_hash.json', 'w') as f:
    json.dump(json_data_dict, f, indent=2)
```
